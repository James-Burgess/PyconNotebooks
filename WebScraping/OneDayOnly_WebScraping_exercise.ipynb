```python
import lxml.html
import requests
import pandas as p
```

```python
URL = 'https://www.onedayonly.co.za/'
```

```python
r = requests.get(URL)
```

```python
doc = lxml.html.fromstring(r.text)
```

```python
deals = doc.cssselect('.product_block')
```

```python
deals
```

```python
deal = deals[0]
```

```python
def parse_deal(deal):
    brand = deal.cssselect('a > span.details > span.titles > h2')[0].text_content()
    name = deal.cssselect('a > span.details > span.titles > p')[0].text_content()
    price = deal.cssselect('a > span.details > span.prices > span.line_two > h3')[0].text_content()
    return (brand, name, price)
parse_deal(deal)
```
